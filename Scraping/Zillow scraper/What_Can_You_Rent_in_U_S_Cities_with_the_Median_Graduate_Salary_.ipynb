## Scrape listings

```python
def scrape_page(url, proxies=None):
    response = requests.get(url, headers=dict_header, proxies=proxies)
    response.raise_for_status()
    time.sleep(max(5.564878, 10.574878 + 2 * np.random.randn()))
    return response

def save_json(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

def get_result_count(soup):
    result_count = soup.find_all('span', 'result-count')[0].text
    return int(result_count.replace(' results', '').replace(',', ''))
```

```python
def get_number_of_pages(result_count):
  number_of_pages = result_count//40
  if result_count%40 > 0:
    number_of_pages+=1
  return number_of_pages
```

```python
for index, row in df[217:218].iterrows():
    url = row['url']
    area_name = row['area_name'].replace('/', '-')

    try:
      response = scrape_page(url, proxies=proxies)
      soup = BeautifulSoup(response.text, 'html.parser')
      listings = soup.find_all('script')
      for listing in listings:
        try:
          json.loads(listing.text)
          if json.loads(listing.text)["props"]:
            data = json.loads(listing.text)
        except:
          continue

      full_file_path = f'{file_path}{area_name}.json'
      save_json(data, full_file_path)

      print(f'Done page 1: {area_name}')

      result_count = get_result_count(soup)
      number_of_pages = get_number_of_pages(result_count)

      if number_of_pages > 1:
        for page in range(2, number_of_pages + 1):
          try:
            pagination = f"pagination%22%3A%7B%22currentPage%22%3A{page}%7D%7D"
            new_url = url.split('pagination')[0] + pagination
            response = scrape_page(new_url, proxies=proxies)
            soup = BeautifulSoup(response.text, 'html.parser')
            listings = soup.find_all('script')
            for listing in listings:
              try:
                json.loads(listing.text)
                if json.loads(listing.text)["props"]:
                  data = json.loads(listing.text)
              except:
                continue
            full_file_path = f'{file_path}{area_name}_page{page}.json'
            save_json(data, full_file_path)
            result_count = get_result_count(soup)
            print(f'Done page {page}: {area_name}')
          except Exception as e:
            print(f'Error with {area_name} page {page}: {e}')

    except Exception as e:
        print(f'Error with {area_name}: {e}')
```

# Read scraped data

```python
json_files = [f for f in os.listdir(file_path) if f.endswith('.json')]
```

```python
data = []
```

```python
for file in json_files:
    with open(os.path.join(file_path, file), 'r') as json_file:
      try:
        json_data = json.load(json_file)
        json_data = json_data['props']['pageProps']['searchPageState']['cat1']['searchResults']['listResults']
        file_name = file.split('.json')[0].split('_')[0]

        for item in json_data:
          id = item.get('id', None)
          price = item.get('unformattedPrice', None)
          size = item.get('area', None)

          data.append({'area_name': file_name, 'id': id, 'price': price,
                       'size': size})

      except Exception as e:
        print(f'Error with {file}: {e}')
```
